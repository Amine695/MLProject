### Question 2 : Nettoyage des données

On veut dans cette question nettoyer notre fichier csv dans le cas où l'on aurait des doublons, des valeurs manquantes, etc ...

#### Suppression des valeurs manquantes

Pour vérifier si nous avons des valeurs manquantes dans notre fichier csv, on rajoute le paramètre **na_values= 'n/a'** dans notre data frame, et ainsi lorsque l'on met **df.head()**, on vérifie si on à des **NaN** signifiant *Not a number*. Par exemple pour les 30 premières lignes on s'apercoit que l'on a des valeurs manquantes: la 10ème musique n'a pas de date de sortie, la 16 ème musique n'as pas de nom, ou encore la 26ème qui n'as pas de loudness.  

```python
missing_values = ["n/a", "nan"]
df = pd.read_csv('data_music.csv',na_values =missing_values) #on rajoute ce paramètre pour qu'il nous indique les valeurs manquantes
#df.head(50)
```

Ainsi, on veut enlever les valeurs manquantes du csv . On peut se servir de la commande dropna() qui se charge de supprimer les lignes contenant une ou plusieurs valeurs manquantes. Chose importante, on met en paramètre **inplace = True** signifiant que l'on veut que les modifications soient appliquées sur le dataframe que l'on utilise actuellement.

```python
df.dropna(inplace= True) #on met le inplace sur True pour modifier notre dataframe d'origine
#df.head(50)
```

On peut vérifier que l'on a bien enlever les valeurs manquantes en tapant la commande **df.info()** où l'on constate que l'on est passé de 178000 lignes à environ 158000 lignes. Faisons maintenant la même chose pour les doublons !

```python
df.info()
```

#### Suppression des doublons

Dans la même idée, on utilise la fonction **drop_duplicates** pour enlever les doublons de notre fichier, puis on vérifie avec un **df.info()** et on remarque que l'on passe de 158000 lignes à 156000 une fois les doublons supprimés. Notre nettoyage de notre dataframe est maintenant terminé et l'on travaillera sur ce dataframe modifié pour la suite du TP

```python
df.drop_duplicates(keep='first',inplace = True)
#df.head(50)
```

Pour vérifier que les changements on bien été pris en compte:

```python
df.info()
```

### Question 1 : Analyse des données

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

On trace ici la **matrice de corélation** qui nous donne des informations sur l'importance de certains attributs par rapport à d'autres. Exemple, on voit que **energy** est fortement dépendant de **loudness** et inversement. On passe maintenant aux tracé de graphiques, d'histogrammes et de boxplots pour essayer d'en apprendre plus sur notre dataset.

Ci dessous l'histogramme indiquant la **proportion des musiques populaires et moins populaires**. On remarque déjà que la majorité des musiques sont peu populaires (<20) , tandis que le reste se situe entre 20 et 60 avec une minorité de musiques ayant une populairté > 60. <br/>On peut en déduire que notre dataset contient des musiques connus et aimé du public. 

```python
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(10,5))
ax = df[['popularity']].plot(kind='hist',bins=[0,20,40,60,80,100],rwidth=0.4,color='#abcdef')
ax.set_xlabel("Popularité de la musique")
ax.set_ylabel("Nombre de musique")
plt.title("Répartition des musiques selon leur popularité",fontdict = {'fontweight':'bold','fontsize':'15'})
plt.show()
```

Sur cet histogramme est référencé le nombre de musiques sortis entre 1920 et 2020. On constate que le nombre de musiques produites n'ont cesser d'augmenter entre **1920 et 1950** malgré la 2nd guerre mondiale, avant de stagner à environ 17000 jusqu'à les années 2010 où l'on constate un bond.<br/> Cela peut s'expliquer notemment par l'apparition des moyens de communication et des réseaux à la fin du XX ème siècle puis ensuite des plateformes de streaming en ligne ainsi que de la rémunération des artistes qui n'a cesser de croître au fil du temps. 

```python
plt.rcParams.update({'font.size': 14})
ax = df['year'].plot(kind='hist',bins=10,rwidth=0.4 ,color="#abcdef")
ax.set_xlabel("années")
ax.set_ylabel("nombre de musiques")
plt.title("Evolution du nombre de musique au cours du temps",fontdict = {'fontweight':'bold','fontsize':'15'})
```

Sur ce graphe, on s'intéresse à la corrélation entre intensité et popularité. On remarque que au plus l'intensité d'une musique est grande, au plus elle a de chance d'être populaire. Au fil du temps, de nouveaux genre musicaux sont apparut comme le hip-pop, le Rock ou encore le métal puis les DJ's avec leur forte basse. Ainsi cette corrélation peut s'expliquer par le fait qu'un son à forte intensité vas éventuellement plus impacter une personne sur son humeur, et ainsi déchainer les foules lors des concerts par exemple.

```python
popularite=np.arange(0,101)
plt.rcParams.update({'font.size': 14})
liste_annee=np.zeros((popularite.size))

for p in popularite:
    df_p = df[ df['popularity'] == p ]

    moyenne=df_p['loudness'].mean()

    liste_annee[p]=moyenne
plt.plot(popularite,liste_annee,'g.-')
plt.ylabel("Intensité de musique")
plt.xlabel("Popularité de la musique")
plt.title("Évolution de la popularité des musiques en fonction de leur intensité",fontdict = {'fontweight':'bold','fontsize':'15'})
plt.show()
```

Ainsi, pour venir corréler et valider notre graphe précedent, on vas réaliser le même graphe avec cette fois-ci l'étude de la dansabilité d'une musique, c'est-à-dire son taux "d'ambiancement" en fonction de la popularité toujours. On remarque que l'on a bien le résultat souhaiter puisqu'on voit qu'au plus une musique ambiance, au plus elle est populaire et pour qu'une musique ambiance, il faut que son intensité soit assez forte raison pour laquelle on a deux courbes similaires pour l'intensité et la dansabilité, chose assez logique je trouve.

```python
popularite=np.arange(0,101)
plt.rcParams.update({'font.size': 14})
liste_annee=np.zeros((popularite.size))

for p in popularite:
    df_p = df[ df['popularity'] == p ]

    moyenne=df_p['danceability'].mean()

    liste_annee[p]=moyenne
plt.plot(popularite,liste_annee,'r.-')
plt.ylabel("Dansabilité d'une musique")
plt.xlabel("Popularité de la musique")
plt.title("Évolution de la popularité des musiques en fonction de leurs dansabilité",fontdict = {'fontweight':'bold','fontsize':'15'})
plt.show()
```

Enfin, si l'on étudie maintenant l'intensité musicale (loudness) en fonction de l'énergie, on a également un résultat très logique entre ces deux attributs. En effet, pour les musiques ayant une faible énergie, leur intensité se situe entre **-50** et **-10** et plus l'énergie augmente et plus cet intervalle diminue jusqu'à atteindre pour les énergie élevées une intensité entre **-20** et **0**. **Conclusion :** **On a une corrélation entre l'intensité, la dansabilité et l'énergie d'une musique.** Plus une musique est intense, plus son taux de dansabilité est élever et plus elle a d'énergie. Ces 3 combinés font qu'elle a de très forte chance d'être populaire. 

```python
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(10,5))
plt.scatter(df['energy'],df['loudness'],c='#abcdef')
plt.xlabel("énergie musicale")
plt.ylabel("intensité musicale")
plt.title("Évolution de l'intensité des musiques en fonction de leurs énergie",fontdict = {'fontweight':'bold','fontsize':'15'})
plt.show()
```

On s'intèresse maintenant à un autre attribut qui est **l'instrumentalité**. Ya t-il un lien entre l'instrumentalité d'une musique et sa popularité ? Et bien à vrai dire oui puisque comme vous pouvez le voir sur ce graphe, plus une musique est instrumentalisé et moins elle a de chance d'être populaire. Pour rappel, **la musique instrumentale caractérise toute musique exclusivement destinée à des instruments.** Ainsi, on peut penser que la musique classique, le concerto, la symphonie, etc... ne sont pas des genres musicaux très écouté du grand public, chose assez cohérente je trouve.

```python
popularite=np.arange(0,101)
plt.rcParams.update({'font.size': 14})
liste_annee=np.zeros((popularite.size))

for p in popularite:
    df_p = df[ df['popularity'] == p ]

    moyenne=df_p['instrumentalness'].mean()

    liste_annee[p]=moyenne
plt.plot(popularite,liste_annee,'b.-')
plt.ylabel("instrumentalité d'une musique")
plt.xlabel("Popularité de la musique")
plt.title("Évolution de la popularité des musiques en fonction de leurs instrumentalité",fontdict = {'fontweight':'bold','fontsize':'15'})
plt.show()
```

Sur cet histogramme, on étudie **l'acousticité** des musiques dans notre dataframe. On peut remarquer que la quasi-totalité des musiques ont une acousticité soit très proche de 0 soit très proche de 1. Le peu de musique restant ont une acousticité entre 0.2 et 0.8. Ainsi, les musiques acoustiques ou non acoustiques font à peu près part égale dans notre dataframe 

```python
plt.rcParams.update({'font.size': 14})
ax = sns.displot(df['acousticness'],kde=False,color='#D33E5D')
ax.set(xlabel='Acousiticité de la musique', ylabel='nombre de musique', title="Répartition des musiques en fonction de leur acousticité")
```

On passe maintenant à un tout autre genre de représentation, le **boxplot**. Le boxplot est un moyen très efficace de comparer plusieurs attributs à conditions qu'ils ont la **même échelle**. Ainsi, ici, on compare l'acousticité, la dansabilité, l'énergie ainsi que la valence, tous situé entre 0 et 1. On a donc une idée du minimum, du maximum, ainsi que de la répartition des valeurs autour de la médiane, chose importante car on peut se faire une idée de si un attribut a des **valeurs serrer** autout de la médiane (comme la dansabilité) ou si au contraire a des **valeurs homogénéiquement répartis** (comme l'acousticité).

En analysant le boxplot, on voit que la médiane des 4 attributs est environ la même (0.5), on voit que si l'acousticité des musiques varient du minimum au maximum, la dansabilité elle, se situe majoritairement entre 0.4 et 0.6. On a quasi les mêmes résultats pour l'énergie et la valence répartit en moyenne entre 0.2 et 0.7.

```python
acoustique = df["acousticness"]
danse = df["danceability"]
energie = df['energy']
val = df['valence']
plt.figure(figsize=(10,5))
labels= ["Acousticité","Dansabilité","énergie","valence"]
plt.rcParams.update({'font.size': 14})
boxes = plt.boxplot([acoustique,danse,energie,val], labels = labels, medianprops = {'color':'r' ,'linewidth':2})

for box in boxes['boxes']:
    box.set(color='b',linewidth = 1.5)

plt.title("Boxplot de 4 attributs avec leur min,max et mediane",fontdict = {'fontweight':'bold','fontsize':'15'})

plt.show()
```

Je me suis intéréssé également à la part des musiques comportant du contenu explicite. Comme cette attribut comporte des valeurs booléenne, (0 : explicite, 1 : implicite), je me suis dis que le mieux serait de faire un camenbert :). On remarque ainsi l'écrasante majorité de chansons avec contenu explicite (plus de 93%) . <br/>Pour rappel, le contenu  **Explicit** est appliqué lorsque les paroles ou le contenu d'une chanson contiennent un ou plusieurs des critères suivants qui pourraient être considérés comme offensants ou inappropriés pour les mineurs:

### Question 3: Régression linéaire  multiple

On cherche maintenant à appliquer une régression linéaire multiple et essayer de comparer les résultats avec ceux obtenus du modèle de forêt aléatoire. <br/> On commence par convertir nos colonnes en tableaux numpy. Ensuite on appelle la fonction **LinearRegression()** et on finit par mettre en place la prédicition.<br/> En calculant donc la **MSE**, on a une valeur d'environ **335**, ce qui est bien au dessus du résultat obtenus par notre modèle de forêt aléatoire.<br/>

```python
x = df2.iloc[:, [0,1,2,4,5]].values # on stock les 5 attributs
y = df2.iloc[:, 3].values # et la popularité
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33)#on sépare les données train/test
regr = LinearRegression() #on appelle la régression linéaire
regr.fit(np.array(x_train).reshape(-1,5), y_train) #on entraine le modèle
preds = regr.predict(np.array(x_test).reshape(-1,5)) #on fait la prédiction
print("MSE: ", metrics.mean_squared_error(y_test, preds)) #on calcul la MSE
```

On cherche a afficher le **résidu**. Un résidu est la distance verticale entre un point de donnée et la droite de régression. Chaque point de donnée a un résidu. Il est positif s'il est au-dessus de la ligne de régression et négatife s'il est en dessous de la ligne de régression. 

Ce graphique ci-dessous montre la distribution de l'erreur. La plupart du temps, il semble que les valeurs soient proches de 0. Pourtant, parfois, les valeurs sont très, très fausses. Par exemple, si une musique a une popularité de 30 et que le modèle prédit une popularité de 40, alors "l'erreur" est de 10, si le modèle prédit une popularité de 10, alors "l'erreur" est de -20. C'est ce qu'on appelle plus comunément le **résidu** ou **l'écart d'ajustement**.

```python
residuals = preds - y_test
plt.rcParams.update({'font.size': 14})
plt.hist(residuals,color="#9C52E4",bins = 10)
plt.xlabel("Distribution de l'erreur")
plt.ylabel("Nombre de musiques")
plt.title("Différence entre la popularité d'une musique et sa popularité prédit ",fontdict = {'fontweight':'bold','fontsize':'20'})

plt.show()
```

Ainsi, on peut résumer les avantages et les inconvénients du modèle forêt aléatoire et régression linéaire dans un tableau:<br/>
 <br/> <center><font size="3"> **<u>Forêt Aléatoire**</font></center>

| <center><font size="2"> Avantages</font> <center> | <center><font size="2"> Inconvénients</font> <center> |
|:--|:--|
| - <font size="2"> Algorithme est très stable<br><br/>- L'algorithme  fonctionne bien lorsque l'on a des attributs numériques<br/> et catégorielles.<br><br/>- Le modèle de forêt aléatoire n'est pas biaisé car il existe plusieurs arbres et chaque arbre est formé sur un sous ensemble de données </font>| - <font size="2"> Un inconvénient majeur: la complexité.<br><br/>- Il nécessite beaucoup plus de ressources de calcul en raison du grand nombre d'arbre de décision réunis.<br/><br>- En raison de leur complexité, ce modèle nécessite beaucoup plus de temps d'entrainement que d'autres modèle </font> |


De même, <center><font size="3"> **<u>Régression linéaire**</font></center>

| <center><font size="2"> Avantages</font> <center> | <center><font size="2"> Inconvénients</font> <center> |
|:--|:--|
| - <font size="2"> La régression linéaire est simple à mettre en œuvre et plus facile à interpréter les coefficients de sortie.<br> <br/>- Modèle très efficace lorsque l'on sait que la relation entre la variable <br/>indépendante et la/les variables dépendante(s) est linéaire.<br> <br/>-Le surajustement peut être réduit par la régularisation, c'est-à-dire que le modèle s'adapte de très près à un ensemble de données et capture  donc <br/>également les données bruyantes, ce qui a un impact négatif sur les performances du modèle et réduit sa précision sur l'ensemble de test. </font>|-<font size="2">Les valeurs aberrantes peuvent avoir des effets énormes sur la régression .<br> <br/>- Ce modèle est sujet au sous-ajustement, c'est-à-dire que le modèle ne parvient pas à capturer correctement les données.<br> <br/>- La régression linéaire suppose que les données sont indépendantes, par conséquent, toute multicolinéarité doit être supprimée avant d'appliquer la régression linéaire. </font> |

**<u>Conclusion:**<br/> Le modèle de Forêt aléatoire semble être plus efficace mais aussi plus complexe et gourmand en ressources tandis que la régression linéaire est simple à implémenter et est un excellent outil pour analyser les relations entre les variables. Cependant, elle n'est pas recommandée pour la plupart des applications pratiques car elle simplifie à l'extrême les problèmes du monde réel en supposant une relation linéaire entre les variables, ce qui n'est pas toujours le cas.

### Question 6 : K-fold cross validation 

Jusqu'à maintenant, nous avons divisé les données en ensembles de données **d'entrainement** et de **validation** et les utilisons respectivement pour la formation et le test du modèle. Cette méthode n'est pas très fiable car les données de d'entrainement et de validation n'ont pas toujours le même type de variation que les données d'origine, ce qui affectera la précision du modèle. <br/> Ainsi, la K-fold cross validation résout ce problème en divisant les données d'entrée en plusieurs groupes au lieu de seulement deux groupes.

<img src="kfold.png"
     alt="Fonctionnement du K-Fold Validation"
     style="float: left; margin-right: 10px;" />

Nous allons utiliser **cross_val_score()** pour évaluer un score par cross validation. Nous allons utiliser trois modèles différents pour l'analyse. Nous trouverons le score pour chaque division, puis prendrons la moyenne pour obtenir le score global. <br/> <br/>
**Attention !!** <br/>J'ai fais le K-Cross validation pour le dataframe **df2** (cf question 2). Donc après avoir compilé les codes de la question 5, vous devez recompiler la question 2, sinon la K-fold validation se fera avec le dataframe df3 contenant uniquement les musiques d'après 1975 !!

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import KFold, StratifiedKFold, cross_val_score
from sklearn import linear_model, tree, ensemble
from sklearn.metrics import median_absolute_error
from sklearn.linear_model import LinearRegression
```

On commence par diviser nos données. J'ai choisi de les diviser en 5 groupes et de mélanger à chaque dois les valeurs de sorte à ce que l'ensemble du dataset soit analysé

```python
#on divise les data en 5 
kf = KFold(n_splits=5,shuffle = True, random_state=0)

cnt = 1
# La méthode split() génère des index pour diviser les données en ensemble d'entraînement et de validation.
for train_index, test_index in kf.split(x, y):
    print(f"Groupe: {cnt}, Données d'entrainement: {len(train_index)}, données de validation:{len(test_index)}")
    cnt += 1
```

On voit bien que l'on a 5 groupes différents avec à chaque fois 2/3 des data pour l'entrainement et 1/3 pour la validation. Le paramètre **shuffle = True** indique que nos groupes sont à chaque fois mélangés, donc on n'a pas 5 groupes avec les même data. 

J'ai implémenté une fonction calculant l'erreur quadratique moyenne, comme l'on vas l'utiliser plusieurs fois par la suite.

```python
#fonction calculant l'erreur quadratique moyenne
def rmse(score):
    rmse = np.sqrt(-score)
    mse = -score
    print("MSE moyenne :",mse)
    print(f'Erreur quadratique moyenne: {"{:.2f}".format(rmse)}')
```

### K-fold en utilisant la régression linéaire

```python
score = cross_val_score(linear_model.LinearRegression(), x, y, cv= kf, scoring="neg_mean_squared_error")
print(f'MSE pour chaque groupe: {-score}')
rmse(score.mean())
```

**Temps de compilation > 10mn** !!!

On veut maintenant la k-cross validation mais en faisant varier le nombre d'arbres dans notre RandomForest Regressor. J'ai donc pris des valeurs relativement petite pour ne pas avoir un temps de compilation énorme.

```python
n_estimators = [10,100,150]
mse = 0
rmse = 0
for count in n_estimators:
    score = cross_val_score(ensemble.RandomForestRegressor(n_estimators= count, random_state= 42), x, y, cv= kf, scoring="neg_mean_squared_error")
    print(f'Nombres arbres: {count}')
    rmse(score.mean())
    mse += score.mean()
    rmse += np.sqrt(score.mean())
    
print("MSE moyenne totale :",mse/3)
print("Erreur quadratique moyenne total :",rmse/3)
```

### Question 7 : Méthode PCA

Dans cette question bonus, on cherche maintenant à appliquer la **PCA**, qui est une méthode de réduction. Cette technique est une technique statistique permettant de convertir des données de grande dimension en données de faible dimension en sélectionnant les attributs les plus importants qui capturent le maximum de variance sur l'ensemble de données. L'attribut qui cause la variance la plus élevée est l'attribut le plus important, et ainsi de suite. <br/> Comme on cherche les 5 meilleurs attributs de notre dataset, voilà tout l'intérêt de cette technique.

```python
from sklearn.decomposition import PCA
```

#### PCA sur notre dataframe de départ

On commence par copier notre dataframe dans un nouveau. J'ai supprimé la colonne **id** qui ne me semblait pas utilile et j'ai laissé le reste.

```python
df4 = df.copy()
```
